```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset with missing values
data = {'Age': [22, 25, np.nan, 35, np.nan, 27, 30, 40],
'Insurance': [100000, 250000, 150000, 350000, 100000, np.nan, 200000, 300000]}

df = pd.DataFrame(data)
print("Original Data:\n", df)

# Option 1: Remove rows with missing values
df_dropped = df.dropna()
print("\nData after dropping missing values:\n", df_dropped)

# Option 2: Fill missing values with mean
df_filled = df.fillna(df.mean(numeric_only=True))
print("\nData after filling missing values with mean:\n", df_filled)
```

```python
#Scatter Plot to Show Relationships : Helps visualize relationships between two numerical variables.
#Scatter plot between total_bill and tip
iris = sns.load_dataset('iris')

# Scatter plot: Sepal Length vs Petal Length
plt.figure(figsize=(8, 6))
plt.scatter(iris['sepal_length'], iris['petal_length'], c='blue', alpha=0.7)

plt.title('Scatter plot of Sepal_Length vs Petal_Length')
plt.xlabel('Sepal_Length')
plt.ylabel('Petal_Length')
plt.grid(True)
plt.show()
```
